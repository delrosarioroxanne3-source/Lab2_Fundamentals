```python
# APPENDIX CELL 1 – REDACTION PROTOCOL

target_name = "DELROSARIO"   # your surname in uppercase
STUDENT_ID = "TUPM-25_0924"

SEED_NUM = int(STUDENT_ID[-4])

redacted_name = ""

vowels = "AEIOU"

for char in target_name:
    if char in vowels:
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += char

print("Original Surname:", target_name)
print("SEED_NUM Used:", SEED_NUM)
print("Redacted Output:", redacted_name)
```

```python
# APPENDIX CELL 2 – FORENSIC SIGNATURE

ENGINEER = "DEL ROSARIO"
STUDENT_ID = "TUPM-25-0924"

SEED_NUM = int(STUDENT_ID[-4])

loops = SEED_NUM + 3

first_letter = ENGINEER[0]

A = ord(first_letter)
B = SEED_NUM
C = loops

verification_hash = (A * B) - C

print("First Letter ASCII:", A)
print("SEED_NUM:", B)
print("Total Loops:", C)
print("Verification Hash:", verification_hash)
```

```python
# APPENDIX CELL 3 – PASSWORD LOOP (SIMULATED INPUTS)

correct_password = "admin123"  # Set the password
test_inputs = ["hello", "1234", "admin123"]  # Predefined attempts
attempts = 0
attempts_list = []

# Loop through simulated inputs
for user_input in test_inputs:
    attempts += 1
    attempts_list.append(user_input)
    
    if user_input != correct_password:
        print(f"Attempt {attempts}: {user_input} → ACCESS DENIED ❌")
    else:
        print(f"Attempt {attempts}: {user_input} → ACCESS GRANTED ✅")
        break

# Summary for assessment
print("\n--- SESSION SUMMARY ---")
print(f"Total Attempts Before Success: {attempts}")
print(f"All Inputs Entered: {attempts_list}")
```
